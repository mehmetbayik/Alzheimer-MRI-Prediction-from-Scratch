```python
import os
import random
import datetime
from itertools import product 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

# import random
# random.seed(42)
# np.random.seed(42)
# np.random.RandomState(42)
# os.environ['TF_DETERMINISTIC_OPS'] = '1' 
```

```python
batch1 = '23:52'
comp1 = '12% CPU'
batch512 = '09:43'
comp512 = '54% CPU'
batch5120 = '11:46'
comp5120 = '99% CPU'

batch1_int = int(str(batch1).split(':')[0])*60 + int(str(batch1).split(':')[1])
batch512_int = int(str(batch512).split(':')[0])*60 + int(str(batch512).split(':')[1])
batch5120_int = int(str(batch5120).split(':')[0])*60 + int(str(batch5120).split(':')[1])

# Plot time and comp for different batch sizes
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(7, 3)

# Plot time
ax[0].bar(['1','512','5120'],[batch1_int/60, batch512_int/60, batch5120_int/60])
ax[0].set_title('Time')
ax[0].set_xlabel('Batch Size')
ax[0].set_ylabel('Time (min)')

# Plot comp
comp_values = [int(comp1.strip('% CPU')), int(comp512.strip('% CPU')), int(comp5120.strip('% CPU'))]
ax[1].bar(['1','512','5120'], comp_values)
ax[1].set_title('Computation Power')
ax[1].set_xlabel('Batch Size')
ax[1].set_ylabel('CPU Usage (%)')

plt.tight_layout()
plt.show()
```

```python

lr512 = '07:46'
svm512 = '08:12'
nn512 = '09:43'
complr = '43% CPU'
compsvm = '68% CPU'
compnn = '83% CPU'

lr512_int = int(str(lr512).split(':')[0])*60 + int(str(lr512).split(':')[1])
svm512_int = (int(str(svm512).split(':')[0])*60 + int(str(svm512).split(':')[1]))/(2.5)
nn512_int = int(str(nn512).split(':')[0])*60 + int(str(nn512).split(':')[1])

# Plot time and computation percent for different models
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(7, 3)

# Plot time
ax[0].bar(['Logistic Regression','SVM','Neural Network'],[lr512_int/60, svm512_int/60, nn512_int/60])
ax[0].set_title('Time for 1000 Epochs')
ax[0].set_xlabel('Model')
ax[0].set_ylabel('Time (min)')

# Plot computation percent
comp_values = [int(complr.strip('% CPU')), int(compsvm.strip('% CPU')), int(compnn.strip('% CPU'))]
ax[1].bar(['Logistic Regression','SVM','Neural Network'], comp_values)
ax[1].set_title('Computation Power')
ax[1].set_xlabel('Model')
ax[1].set_ylabel('CPU Usage (%)')

plt.tight_layout()
plt.show()
```
